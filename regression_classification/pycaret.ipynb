```python
import pandas as pd

from pycaret.regression import *
```

```python
df = pd.read_csv('illinois-home-prices.csv')
df = df[['average_home_price', 'pct_college_education', 'unemployment_rate', 'median_household_income']]
```

```python
s = setup(df, target = 'average_home_price', session_id = 0)
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```
