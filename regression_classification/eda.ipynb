# **Exploratory Data Analysis**

## Imports

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('illinois-home-prices.csv')
```

-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

## `average_home_price`

```python
sns.boxplot(data = df, y = 'average_home_price')
plt.ylabel('Average Home Price ($)')
plt.title('Boxplot of Average Home Price by County')
plt.show()

df['average_home_price'].hist()
plt.xlabel('Average Home Price ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Average Home Price by County')
plt.show()
```

```python
print('Mean: ', df['average_home_price'].mean())
print('Median: ', df['average_home_price'].median())

print('Varaince: ', df['average_home_price'].var())
print('Standard Deviation: ', df['average_home_price'].std())
print('Range: ', df['average_home_price'].max() - df['average_home_price'].min())
IQR = df['average_home_price'].quantile(0.75) - df['average_home_price'].quantile(0.25)
print('IQR: ', IQR)

print('Skewness: ', df['average_home_price'].skew())
print('Kurtosis: ', df['average_home_price'].kurtosis())

outliers = df[(df['average_home_price'] < 1.5 * IQR - df['average_home_price'].quantile(0.25)) | (df['average_home_price'] > 1.5 * IQR + df['average_home_price'].quantile(0.75))].shape[0]
print('Number of Outliers: ', outliers)
print('Prevalence of Outliers: ', outliers / df.shape[0])
```

### Exploring Correlation with the Response Variable ( `average_home_price` ) and Multicollinearity

`median_household_income` has the greatest correlation coefficient with `average_home_price` at 0.8828. Next, `pct_college_education` with 0.7559 and `unemployment_rate` with -0.1645. Therefore, the median household income and percentage of the population with a college education are highly correlated with a county's average home price.

`pct_college_education` and `median_household_income` is the only explanatory variable pair with a significant correlation coefficient with each other at 0.6789. Therefore, the percentage of the population with a college education and the median household income of counties exhibit a collinear relationship. During model construction, numerous reglarization techniques will be utilized to counteract the negative effects of multicollinearity if they are present.

There is no apparent three-dimensional correlation between any pair of the explanatory variables and the response variable.
