# **Data Analysis**

## Imports

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('illinois-home-prices.csv')
```

## `average_home_price`

```python
print('`average_home_price`:')

print(df['average_home_price'].head())
```

```python
sns.boxplot(data = df, y = 'average_home_price')

plt.ylabel('Average Home Price ($)')

plt.title('Boxplot of Average Home Price by County')

plt.show()

df['average_home_price'].hist()

plt.xlabel('Average Home Price ($)')

plt.ylabel('Frequency')

plt.title('Distribution of the Average Home Price by County')

plt.show()
```
